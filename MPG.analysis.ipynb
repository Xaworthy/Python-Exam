# Create a visual showing the relationship between horsepower and fuel economy.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data - horsepower (HP) and miles per gallon (MPG)
data = {'Horsepower': [120, 150, 200, 250, 300, 350, 400, 450, 500],
    'Fuel Economy (MPG)': [30, 28, 25, 22, 20, 18, 16, 15, 13]}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Horsepower", y="Fuel Economy (MPG)", s=100, color="b")

# Add titles and labels
plt.title("Relationship between Horsepower and Fuel Economy", fontsize=16)
plt.xlabel("Horsepower", fontsize=14)
plt.ylabel("Fuel Economy (MPG)", fontsize=14)

# Show the plot
plt.show()
```

# -The more Horsepower required the less mgp you will have avalible

# Create a table summarizing fuel economy by country of origin

```python
# Sample data
data = {'Country': ['USA', 'USA', 'Japan', 'Japan', 'Germany', 'Germany', 'USA', 'Japan', 'Germany'],
    'Fuel Economy (MPG)': [20, 22, 35, 32, 28, 30, 19, 36, 27]}

# Group by 'Country' and calculate summary statistics for 'Fuel Economy (MPG)'
summary_table = df.groupby('Country')['Fuel Economy (MPG)'].agg(
    Average_MPG='mean',
    Median_MPG='median',
    Max_MPG='max',
    Min_MPG='min'
).reset_index()

# Display the summary table
print(summary_table)
```

# - Looks like japan gives you the best mgp overall.

```python
# Sample data - average fuel economy by year
data = {'Year': [2000, 2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018, 2020],
    'Average Fuel Economy (MPG)': [20, 21, 22, 22, 23, 24, 25, 26, 27, 28, 29]}

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Year", y="Average Fuel Economy (MPG)", marker="o", color="b")

# Add titles and labels
plt.title("Change in Average Fuel Economy Over Time", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Fuel Economy (MPG)", fontsize=14)

# Show the plot
plt.show()
```

## - From the year 2000 going foward, there has been positive progession producing more mpg for consumers each year

# Create two other visuals or tables related to fuel economy.
